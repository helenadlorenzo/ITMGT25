```python
# 1. Positives and Negatives
    
Positive = 0
Negative = 0
Zeroes = 0

while True: 
    guess = input("Enter input: ")
    if guess=="a":
        break
    elif int(guess)<0:
        Negative += 1
        pass
    elif int(guess)==0:
        Zeroes += 1
    elif int(guess)>0:
        Positive += 1

print(f"Positive: {Positive}")
print(f"Negative: {Negative}")
print(f"Zeroes: {Zeroes}")
```

```python
# 2. The Four-Six Pur Over Coffee Brewing Method

# water = 15 x coffee
# 20 g coffee = 300 g water
# 30 g coffee = 450 g water
# 40 g coffee = 600 g water

# input amount of coffee in grams
# only accept valid inputs
# print total amount of water

coffee = int(input("enter amount of coffee in grams: "))
water = coffee * 15

while True:
    balance = input("choose between sweet or acidic: ")
    if balance=="sweet":
        p1 = water * .40
        p1_1 = water * .16
        p1_2 = water * .24
        break
    elif balance=="acidic":
        p1 = water * .40
        p1_1 = water * .24
        p1_2 = water * .16
        break
    else:
        print("Try again.")
        continue

while True:
    strength = input("Enter Y or y for strong or N or n for not strong: ")
    if strength=="Y" or strength=="y":
        w_p2 = water * .60
        p2 = (water * .60)/4
        break
    elif strength=="N" or strength=="n":
        w_p2 = water * .60
        p2 = (water * .60)/2
        break
    else:
        print("Try again.")
        continue
        
print("Water pour plan:")
print(f"First Part {p1}g:")
print(f"pour #1: {p1_1}")
print(f"pour #2: {p1_2}")
print(f"Second Part {w_p2}g:")
print(f"pour #3: {p2}")
print(f"pour #4: {p2}")
print(f"pour #5: {p2}")
print(f"pour #6: {p2}")
```
